## Reflection

Our plan uses lists to store the deck, the player’s hand, and the dealer’s hand, which makes the game easy to manage. We also use a tuple in the `initial_deal()` function to return both hands at the same time.

After comparing our functions with another group, we simplified our design so that each function has only one clear purpose. For example, `create_deck()` builds and shuffles the deck, while `player_turn()` and `dealer_turn()` only handle their own phases.

These changes make the code easier to understand, reduce errors, and keep the game flow clear and organized.
